```
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```
n_inputs = 40
n_hidden1 = 600
n_hidden2 = 500
n_hidden3 = 400
n_hidden4 = 300
n_hidden5 = 400
n_hidden6 = 300
n_hidden7 = 300
n_hidden8 = 200
n_hidden9 = 200
n_hidden10 = 100
n_outputs = 2000

drop_out = 0.3

reset_graph()
with tf.name_scope("inputs"):
  X1 = tf.placeholder(tf.float32,shape=[None,n_inputs])
  y1 = tf.placeholder(tf.int64,shape=[None])
  training = tf.placeholder_with_default(False, shape=(), name='training')


with tf.name_scope("layers"):
  hidden1 = tf.layers.dense(X1,n_hidden1,activation = tf.nn.relu)
  hidden1_dropout = tf.layers.dropout(hidden1,drop_out,training= training)

  hidden2 = tf.layers.dense(hidden1_dropout,n_hidden2,activation = tf.nn.relu)
  hidden2_dropout = tf.layers.dropout(hidden2,drop_out,training= training)

  hidden3 = tf.layers.dense(hidden2_dropout,n_hidden3,activation = tf.nn.relu)
  hidden3_dropout = tf.layers.dropout(hidden3,drop_out,training= training)

  hidden4 = tf.layers.dense(hidden3_dropout,n_hidden4,activation = tf.nn.relu)
  hidden4_dropout = tf.layers.dropout(hidden4,drop_out,training= training)

  hidden5 = tf.layers.dense(hidden4_dropout,n_hidden5,activation = tf.nn.relu)
  hidden5_dropout = tf.layers.dropout(hidden5,drop_out,training= training)

  hidden6 = tf.layers.dense(hidden5_dropout,n_hidden6,activation = tf.nn.relu)
  hidden6_dropout = tf.layers.dropout(hidden6,drop_out,training= training)

  hidden7 = tf.layers.dense(hidden6_dropout,n_hidden7,activation = tf.nn.relu)
  hidden7_dropout = tf.layers.dropout(hidden7,drop_out,training= training)

  hidden8 = tf.layers.dense(hidden7_dropout,n_hidden8,activation = tf.nn.relu)
  hidden8_dropout = tf.layers.dropout(hidden8,drop_out,training= training)

  hidden9 = tf.layers.dense(hidden8_dropout,n_hidden9,activation = tf.nn.relu)
  hidden9_dropout = tf.layers.dropout(hidden9,drop_out,training= training)

  hidden10 = tf.layers.dense(hidden9_dropout,n_hidden10,activation = tf.nn.relu)
  hidden10_dropout = tf.layers.dropout(hidden10,drop_out,training= training)

with tf.name_scope("logits"):
  logits = tf.layers.dense(hidden4_dropout,n_outputs)

with tf.name_scope("loss"):
  xentrophy = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=logits,labels=y1)
  loss = tf.reduce_mean(xentrophy)

with tf.name_scope("optimize"):
  optimizer = tf.train.AdamOptimizer(learning_rate=0.01)
  training_ops = optimizer.minimize(loss)

with tf.name_scope("accuracy"):
  correct = tf.nn.in_top_k(logits,y1,1)
  accuracy = tf.reduce_mean(tf.cast(correct,tf.float32))

init = tf.global_variables_initializer()
```

```
def next_batch(num, data, labels):
    '''
    Return a total of `num` random samples and labels. 
    '''
    idx = np.arange(0 , len(data))
    np.random.shuffle(idx)
    idx = idx[:num]
    data_shuffle = [data[ i] for i in idx]
    labels_shuffle = [labels[ i] for i in idx]

    return np.asarray(data_shuffle), np.asarray(labels_shuffle)
```

```
n_epochs = 10
batch_size = 66

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(5):
            X_batch, y_batch =next_batch(batch_size,X_train.values,y_train.values)
            #print(X_batch.shape,y_batch.shape)
            sess.run(training_ops, feed_dict={training: True, X1: X_batch, y1: y_batch})
        acc_train = accuracy.eval(feed_dict={X1: X_batch, y1: y_batch})
        acc_test = accuracy.eval(feed_dict={X1: X_test, y1: y_test})
        print("Epoch", epoch, "Train accuracy =", acc_train, "Test accuracy =", acc_test)
```
